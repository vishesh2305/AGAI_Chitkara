### Part A - Architecture Code (Keras)

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense
```

```python
TIMESTEPS = 10
FEATURES=4
HIDDEN_UNITS=3
```

```python
model_rnn = Sequential([
    SimpleRNN(units=HIDDEN_UNITS,input_shape=(TIMESTEPS, FEATURES)),
    Dense(1, activation='sigmoid')
])
```

```python
model_lstm = Sequential([
    LSTM(units=HIDDEN_UNITS, input_shape=(TIMESTEPS, FEATURES)),
    Dense(1, activation='sigmoid')
])
```

```python
print("Simple RNN Model")
model_rnn.summary()
```

```python
print("LSTM Model")
model_lstm.summary()
```
