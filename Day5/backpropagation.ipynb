```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
def sigmoid_derivative(x):
    return x *(x-1)
```

```python
def linear(x):
    return x
```

```python
def layer_forward(inputs, weights, biases, activation):
    outputs=[]
    for neuron_weights, neuron_bias in zip(weights,biases):
        weighted_sum = sum(i*w for i, w in zip(inputs, neuron_weights))+neuron_bias
        outputs.append(activation(weighted_sum))
    return outputs
```

```python
def mse(predicted, actual):
    return sum((p-a) ** 2 for p, a in zip(predicted, actual))/len(predicted)
```

```python
def random_weights(n_neurons, n_inputs):
    return [[random.uniform(-0.1, 0.1) for _ in range(n_inputs)] for _ in range(n_neurons)]
```

```python
def random_baises(n_neurons):
    return [random.uniform(-0.1, 0.1) for _ in range(n_neurons)]
```

```python
def update_weights(weights, gradients, lr):
    return [
        [weights[i][j] - lr * gradients[i][j] for j in range(len(weights[i]))]
        for i in range(len(weights))
    ]
```

```python
def update_biases(biases, gradients, lr):
    return [biases[i] - lr* gradients[i] for i in range(len(biases))]
```

```python
def backward_pass(inputs, hidden_output, raw_output, expected, hidden_weights, output_weights):
    output_gradients = [raw_output[i] - expected[i] for i in range(len(raw_output))]
    output_weight_grads = [[grad * h for h in hidden_output] for grad in output_gradients]
    output_bias_grads = output_gradients

    hidden_gradients=[]
    for j in range(len(hidden_output)):
        grad = sum(output_gradients[i] * output_gradients[i][j] for i in range(len(output_gradients)))
        grad*= sigmoid_derivative(hidden_output[j])
        hidden_gradients.append(grad)

    hidden_weight_grads = [[grad * x for x in inputs] for grad in hidden_gradients]
    hidden_bias_grads = hidden_gradients
    return output_weight_grads, output_bias_grads, hidden_weight_grads, hidden_bias_grads
```
